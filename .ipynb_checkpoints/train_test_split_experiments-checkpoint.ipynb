```python
import numpy as np
from sklearn.model_selection import GroupKFold
import scipy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from IPython import InteractiveShell
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

InteractiveShell.ast_node_interactivity = 'all'
```

```python
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y = np.array([1, 2, 3, 4])
groups = np.array([0, 0, 2, 2])
group_kfold = GroupKFold(n_splits=2)

train_index, test_index = next(group_kfold.split(X, y, groups))
```

```python
#np.random.seed(260)
InteractiveShell.ast_node_interactivity = 'none'
```

```python
results_df = pd.DataFrame()

for true_prob in [0.07]:
    #true_prob=0.3
    n_samples = 10000
    y = scipy.stats.bernoulli(true_prob).rvs((n_samples,1))
    X = np.concatenate([y,y,y,y], axis=1)
    n, m = X.shape
    noise = 2 * np.random.randn(n,m)
    X = X + noise
    X = pd.DataFrame(X)
    y = pd.Series(y.flatten())

    for i in range(300):
    
        stratify = False

        if stratify:
            x_train, x_test, y_train, y_test = train_test_split(
                X, y, stratify=y, test_size=0.1, random_state=i
            )
        else:
            x_train, x_test, y_train, y_test = train_test_split(
                X, y, test_size=0.1
            )

        train_deviation = (
            true_prob - y_train.value_counts(normalize=True).reindex([0,1]).loc[1]
        ) / true_prob
        test_deviation  = (
            true_prob - y_test.value_counts(normalize=True).reindex([0,1]).loc[1]
        ) / true_prob

        lr = LogisticRegression(solver='liblinear', class_weight='balanced');
        lr.fit(x_train, y_train);
        proba_train = lr.predict_proba(x_train)[:,1]
        proba_test  = lr.predict_proba(x_test)[:,1]
        train_auc = roc_auc_score(y_score=proba_train, y_true=y_train)
        test_auc  = roc_auc_score(y_score=proba_test, y_true=y_test)

        # Set distinction task
        train_label = pd.Series(np.full(shape=x_train.shape[0], fill_value=0))
        test_label  = pd.Series(np.full(shape=x_test.shape[0], fill_value=1))
        set_labels  = pd.concat([train_label,test_label])

        set_data = pd.concat(
            [pd.concat([x_train, x_test]),
             pd.concat([y_train, y_test])], 
            axis=1)

        set_data_train, set_data_test, set_labels_train, set_labels_test = train_test_split(
                set_data, set_labels, stratify=set_labels, test_size=0.2
        )

        lr = LogisticRegression(solver='liblinear', class_weight='balanced');
        lr.fit(set_data_train, set_labels_train);
        set_proba_train = lr.predict_proba(set_data_train)[:,1]
        set_proba_test  = lr.predict_proba(set_data_test)[:,1]
        set_train_auc = roc_auc_score(y_score=set_proba_train, y_true=set_labels_train)
        set_test_auc  = roc_auc_score(y_score=set_proba_test,  y_true=set_labels_test)

        result = {
            'true_prob': [true_prob],
            'n_samples': [n_samples],
            'stratify': [stratify],
            'train_deviation': [train_deviation],
            'test_deviation': [test_deviation],
            'train_auc': [train_auc],
            'test_auc': [test_auc],
            'perf_drop': [(train_auc-test_auc)/train_auc],
            'set_train_auc': [set_train_auc],
            'set_test_auc': [set_test_auc],
            'set_perf_drop': [(set_test_auc-set_train_auc)/set_train_auc]
        }
        #print(results_df)

        results_df = results_df.append(pd.DataFrame(result))
```
