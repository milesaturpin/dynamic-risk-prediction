```python
def batch_generator(X, y, batch_size):
    total_batches = len(X) // batch_size
    current_batch = 0
    while True:
        start = batch_size * current_batch
        end = start + batch_size
        yield (X[start:end], y[start:end])
        current_batch = (current_batch + 1) % total_batches

training_generator = batch_generator(x_train_scaled, y_train.reshape(-1,1), batch_size=10)
# Later, call next(training_generator) to get a new batch of the form (X, y)
```

```python
def loss(betas, x, y, C):
    pred = tf.sigmoid(tf.matmul(x, betas))
    y    = tf.constant(y, dtype=tf.float64)
    err  = tf.nn.sigmoid_cross_entropy_with_logits(logits=pred, labels=y) 
    #+ C*tf.norm(betas)
    return tf.reduce_mean(err)
```

```python
def risk_set(data_t):
    size = len(data_t)
    risk_set = np.zeros(shape=(size, size))
    for idx in range(size):
        temp = np.zeros(shape=size)
        t_i = data_t[idx]
        at_risk = data_t > t_i
        temp[at_risk] = 1
        risk_set[idx] = temp
    return risk_set

def batch_generator(X, surv, batch_size):
    total_batches = len(X) // batch_size
    current_batch = 0
    while True:
        start = batch_size * current_batch
        end = start + batch_size
        x_batch, t_batch, e_batch = X[start:end], surv[start:end,0], surv[start:end,1]
        yield (x_batch, t_batch, e_batch)
        current_batch = (current_batch + 1) % total_batches

training_generator = batch_generator(
    x_train.values, surv_train.values, batch_size=60
)

def c_index(betas, x, time_of_event, event_ind):
    """
    Calculate c-index by comparing X @ beta values, and counting
    correct number of rankings.
    """
    observed_event_set   = np.where(event_ind == 1)[0]
    unobserved_event_set = np.where(event_ind == 0)[0]
    correct_ranking_count, num_comparisons = 0, 0
    
    # Ranking by linear combination of betas w/ X
    hazard_pred = x @ betas
    obs_hazard_preds = hazard_pred[observed_event_set]
    observed_event_times = time_of_event[observed_event_set]
    
    for obs_haz_pred, event_time in zip(obs_hazard_preds, observed_event_times):
        risk_idx   = np.where(time_of_event >= event_time)[0]
        risk_preds = hazard_pred[risk_idx]
        correct_ranking_count += np.sum(obs_haz_pred > risk_preds)
        num_comparisons += risk_idx.shape[0]
        
    return correct_ranking_count / num_comparisons

def survival_log_likelihood(betas, X, idx_observed, risk_matrix, C=1):
    """
    betas - (m,1)
    X - (n, m)
    idx_observed - (e=1,)
    risk_matrix - (n,n), R_ij = 1 if t_j > t_i 
    """
    h = tf.matmul(X, betas)
    exp_h = tf.exp(h)
    regularizer = C * tf.norm(betas)
    return -tf.reduce_mean(
        tf.gather(h, idx_observed) - 
        tf.log(tf.matmul(
            risk_matrix[idx_observed], 
            exp_h
        ))
    ) + regularizer
```

```python
np.random.seed(3531)
betas = tfe.Variable(np.random.randn(num_features, 1), dtype=tf.float64)
betas
optimizer = tf.train.AdamOptimizer(learning_rate=1)
grad      = tfe.implicit_gradients(survival_log_likelihood)
num_iter  = 60
batch_losses = []
batch_c_indexes = []
train_c_indexes = []
test_c_indexes = []
for i in range(1,num_iter+1):
    
    x_batch, t_batch, e_batch = next(training_generator)
    risk_batch = risk_set(t_batch)
    idx_observed = np.where(e_batch == 1)
    optimizer.apply_gradients(grad(betas, x_batch, idx_observed, risk_batch))
    
    # Print out loss and c-index
    batch_loss = survival_log_likelihood(betas, x_batch, idx_observed, risk_batch).numpy()
    batch_c_index = c_index(betas.numpy(), x_batch, t_batch, e_batch)
    train_c_index = c_index(
        betas.numpy(), x_train.values, surv_train.values[:,0], surv_train.values[:,1]
    )
    test_c_index = c_index(
        betas.numpy(), x_test.values, surv_test.values[:,0], surv_test.values[:,1]
    )
    batch_losses.append(batch_loss)
    batch_c_indexes.append(batch_c_index)
    train_c_indexes.append(train_c_index)
    test_c_indexes.append(test_c_index)
    if i % (num_iter//10) == 0:
        print(f'Iter {i}, loss = {batch_loss}, c_index = {batch_c_index}')
betas
```

```python
def risk_set(data_t):
    size = len(data_t)
    risk_set = np.zeros(shape=(size, size))
    for idx in range(size):
        temp = np.zeros(shape=size)
        t_i = data_t[idx]
        at_risk = data_t > t_i
        temp[at_risk] = 1
        risk_set[idx] = temp
    return risk_set

def batch_generator(X, surv, batch_size):
    total_batches = len(X) // batch_size
    current_batch = 0
    while True:
        start = batch_size * current_batch
        end = start + batch_size
        x_batch, t_batch, e_batch = X[start:end], surv[start:end,0], surv[start:end,1]
        yield (x_batch, t_batch, e_batch)
        current_batch = (current_batch + 1) % total_batches

training_generator = batch_generator(
    x_train.values, surv_train.values, batch_size=60
)

def c_index(betas, x, time_of_event, event_ind):
    """
    Calculate c-index by comparing X @ beta values, and counting
    correct number of rankings.
    """
    observed_event_set   = np.where(event_ind == 1)[0]
    unobserved_event_set = np.where(event_ind == 0)[0]
    correct_ranking_count, num_comparisons = 0, 0
    
    # Ranking by linear combination of betas w/ X
    hazard_pred = x @ betas
    obs_hazard_preds = hazard_pred[observed_event_set]
    observed_event_times = time_of_event[observed_event_set]
    
    for obs_haz_pred, event_time in zip(obs_hazard_preds, observed_event_times):
        risk_idx   = np.where(time_of_event >= event_time)[0]
        risk_preds = hazard_pred[risk_idx]
        correct_ranking_count += np.sum(obs_haz_pred > risk_preds)
        num_comparisons += risk_idx.shape[0]
        
    return correct_ranking_count / num_comparisons

def survival_log_likelihood(betas, X, idx_observed, risk_matrix, C=1):
    """
    betas - (m,1)
    X - (n, m)
    idx_observed - (e=1,)
    risk_matrix - (n,n), R_ij = 1 if t_j > t_i 
    """
    h = tf.matmul(X, betas)
    exp_h = tf.exp(h)
    regularizer = C * tf.norm(betas, 2)
    return -tf.reduce_mean(
        tf.gather(h, idx_observed) - 
        tf.log(tf.matmul(
            risk_matrix[idx_observed], 
            exp_h
        ))
    ) + regularizer
```

```python
def survival_log_likelihood(betas, X, surv):
    # Dot product between betas and data 
    h = tf.matmul(X, betas)
    exp_h = tf.exp(h)
    # Get index of data with observed events
    observed_event_idx = tf.where(
        tf.equal(
            surv[:,1], 
            tf.constant(1, dtype=tf.int64)
        )
    )
    
    #print(h)
    #print(h.shape)
    #print(exp_h)
    #print(observed_event_idx)
    
    log_likelihood = 0 
    # Iterate through data with observed events 
    for idx in observed_event_idx.numpy():
        idx = idx.item()
        # Get xs, dot product, and exp(dot_product) for idx
        xi, hi, exp_hi = (
            tensor[idx] for tensor in [X, h, exp_h]
        )
        # Find all xi's with event times past this one
        event_time = surv[idx,0]
        risk_population = tf.where(surv[:,0] >= event_time)
        # Calculate log likelihood
        log_likelihood = (
            log_likelihood 
            + hi 
            - tf.log(tf.reduce_sum(
                tf.gather_nd(exp_h, risk_population)
            ))
        )
        
        #print(xi,hi,exp_hi)
        #print(event_time)
        #print(risk_population)
        #print(log_likelihood)
        
    return -1*log_likelihood


def logistic_log_likelihood(betas, X, surv):
    # Dot product between betas and data 
    pred = tf.matmul(X, betas)
    arrest = tf.to_double(tf.reshape(surv[:,1], [-1,1]))
#     print(pred.dtype, arrest.dtype)
#     print(betas.shape, X.shape, surv.shape)
#     print(pred.shape, arrest.shape)
#     print(arrest, pred, tf.multiply(arrest, pred))
    
    
    
    v1 = tf.exp(tf.multiply(arrest, pred))
    v2 = tf.add(1., v1)
    v3 = tf.log(v2)
    v33 = tf.nn.sigmoid_cross_entropy_with_logits(logits=pred, labels=arrest) 
    #print(v3, v33)
    #print(v1,v2,v3)
    
    log_pred = tf.log(tf.add(1., tf.multiply(arrest, pred)))
    #print(log_pred.shape)
    
    #print(tf.reduce_sum(v3))
    return -tf.reduce_sum(v3)
```
