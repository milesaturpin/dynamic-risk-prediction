```python
def create_rolling_matrix(data, window):
    """
    Return rolling matrix of shape (m, n, window).
    """
    n, m = data.shape
    buffer = np.zeros((window-1, m))
    s2 = np.concatenate([buffer, data])
    rolling_matrix = np.zeros((m, n, window))
    for i in range(n):
        start = i
        end   = i + window
        # (m,0,window) <- (window,m).T
        rolling_matrix[:,i,:] = s2[start:end].T
    return rolling_matrix

def expsum(data, zerolife):
    n, m = data.shape
    win = 6*zerolife
    expvals = np.array([0.05**(t/zerolife) for t in np.arange(win)][::-1]).reshape((-1,1))
    rolling_matrix = create_rolling_matrix(data, win)
    expsum = rolling_matrix @ expvals
    return expsum[:,:,0].T
```

```python
def create_rolling_matrix(data, window):
    """
    Return rolling matrix of shape (m, n, window).
    Pad with zeros so that we don't shrink.
    """
    n, m = data.shape
    buffer = np.zeros((window-1, m))
    s2 = np.concatenate([buffer, data])
    rolling_matrix = np.empty((m, n, window))
    for i in range(n):
        start = i
        end   = i + window
        # (m,0,window) <- (window,m).T
        rolling_matrix[:,i,:] = s2[start:end].T
    return rolling_matrix

def expsum(data, zerolifes):
    """
    Vectorized form of exponentially weighted sum.
    We truncate the exponential sum when the weight
    reaches 0.05 and we parameterize by the number of
    steps to get there, the "zerolife". 
    
    Params
    ------
    data (np.ndarray) -- matrix of size (n,m)
    zerolifes (list) -- list of length z,
        a zerolife of 3 will have a window_len of 3+1
    
    Return matrix of size (n, m*z)
    """
    z = len(zerolifes)
    n, m = data.shape
    # Take max of zerolifes + 1 to get window length
    max_zerolife = max(zerolifes)
    window_len   = max(zerolifes) + 1
    rolling_matrix = create_rolling_matrix(data, window_len)
    
    # Create matrix of shape (window_len, z)
    # Buffer with 0 to accommodate diff win lengths  
    expvals = np.empty((window_len, z))
    for zi, zerolife in enumerate(zerolifes):
        buffer_len = max_zerolife - zerolife
        expvals[:,zi] = np.array(
            [0] * buffer_len  
            + [ 0.05**(t/zerolife) for t in np.arange(zerolife + 1) ][::-1]
        )

    # (m, n, window_len) @ (window_len, z) -> (m, n, z)
    expsum = rolling_matrix @ expvals
    # Return (n, m*z), flatten multiple computations to 2D
    return np.swapaxes(expsum, 0, 1).reshape((n, m*z))
```

```python
#%%timeit
n,m = 10000,3

np.random.seed(5)
test_data = pd.DataFrame(
    bernoulli.rvs(p=0.05, size=n*m).reshape((n,m))
)

expsum(test_data, zerolifes=[2, 5, 6])
```

```python
%%timeit
expsum(test_data, zerolifes=[6, 24, 48, 144])
```

```python
n, m = 60, 3
zerolifes = [3, 6, 12, 24]
z = len(zerolifes)

fig, axes = plt.subplots(m,1,figsize=(12,m*3))
test_data = pd.DataFrame(
    bernoulli.rvs(p=0.2, size=n*m).reshape((n,m))
)
exp_data = pd.DataFrame(expsum(test_data.values, zerolifes=zerolifes))

for i in range(m):
    test_data.reset_index().plot(
        kind='scatter', x='index', y=i, ax=axes[i]
    )
    exp_data.iloc[:, i*z:(i+1)*z].plot(ax=axes[i])
plt.show();
```
