```python
from sklearn.metrics import roc_auc_score
import pandas as pd

def auc_vs_lambda(fit, x, y):
    # Note: reversed direction so that lambda goes lo -> hi
    
    preds = x @ np.flip(glmnetCoef(fit), axis=1)
    calc_auc = lambda s: roc_auc_score(y_true=y[:,1], 
                                       y_score=s)
    aucs = np.apply_along_axis(arr=preds, axis=0, func1d=calc_auc)
    df = pd.DataFrame(
        {'log_lambda': np.flip(np.log(fit['lambdau'])), 
         'auc':aucs}
    ).set_index('log_lambda').auc
    
    return df

auc_vs_lambda(fit, x_train.values, y_train.values).head()
auc_vs_lambda(fit, x_test.values, y_test.values).head()
```

## Manual Cox CV

1. Create splits for folds, loop through folds
2. Do GLM fit with training data 
3. Pick best lambda? and get performance

```python
import seaborn as sns

num_folds, num_lambdas, num_alphas  = 10, 50, 20
foldid = scipy.random.choice(num_folds, size = y_train.shape[0], replace = True)
alphas = np.round(np.linspace(0,1,num_alphas), 2)
lambdas = np.array([np.exp(x) for x in np.linspace(-6,4,num_lambdas)])

def inner_cv_cox(x, y, foldid):
    perf_arr = np.zeros((num_folds, num_lambdas, num_alphas))
    for i in range(num_folds):
        which = foldid == i
        x_train = x[~which,]
        y_train = y[~which,]
        x_test = x[which,]
        y_test = y[which,] 
        
        for j, alpha in enumerate(alphas):
            fit = glmnet(x=x_train, y=y_train, family='cox', 
                         alpha=alpha, 
                         lambdau=lambdas, 
                         intr=False)
            auc_df = auc_vs_lambda(fit=fit, x=x_train, y=y_train)
            perf_arr[i,:,j] = auc_df.values
    return perf_arr
        
def plot_cv_results(perf_arr, lambdas, alphas):
    """Plot AUC vs. Lambda values across runs for diff alphas."""
    grid_shape = int(np.ceil(num_alphas/2)), 2
    fig, axes = plt.subplots(*grid_shape, figsize=(8, 1.5*num_alphas), 
                             sharey=True, sharex=True)
    log_lambdas = np.round(np.log(lambdas), 1)
    for k in range(num_alphas):
        ax_idx = np.unravel_index(indices=k, dims=grid_shape)
        auc_vs_lambda_df = pd.DataFrame(perf_arr[:,:,k].T,
                                        index=log_lambdas)
        axes[ax_idx].plot(auc_vs_lambda_df.mean(axis=1))
        axes[ax_idx].fill_between(auc_vs_lambda_df.index,
                                  auc_vs_lambda_df.min(axis=1),
                                  auc_vs_lambda_df.max(axis=1),
                                  alpha=0.1)
        axes[ax_idx].set(title=f'alpha={alphas[k]}', xlabel='lambdas', ylabel='auc')
        #axes[ax_idx].set_xticks(log_lambdas)
    plt.tight_layout()
    
    # Heatmap
    plt.figure(figsize=(10,8))
    g = sns.heatmap(np.mean(perf_arr, axis=0))
    g.set(yticklabels=np.round(lambdas, 3))
    for tick in g.get_yticklabels():
        tick.set_rotation(0)
    g.set(xticklabels=np.round(alphas, 2));
    
perf_arr = inner_cv_cox(x_train.values.astype(np.float64),
      y_train.values.astype(np.float64), foldid)
```

```python
plot_cv_results(perf_arr, lambdas, alphas)
```

```python
def get_best_hparams(perf_arr, lambdas, alphas):
    perf_mean = perf_arr.mean(axis=0)
    max_idx = np.unravel_index(perf_mean.argmax(), perf_mean.shape)
    best_lambda = lambdas[max_idx[0]]
    best_alpha  = alphas[max_idx[1]]
    return best_lambda, best_alpha

get_best_hparams(perf_arr, lambdas, alphas)
```
